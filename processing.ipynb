```python
df = pd.read_csv("SlateGunDeaths.csv")
```

```python
df_dropped = df.drop(['date', 'name', 'age', 'url'], axis=1)
```

```python
male = df_dropped[df_dropped['gender'] == 'M']
female = df_dropped[df_dropped['gender'] == 'F']
```

```python
male.head()
female.head()
```

```python
male_by_city = male.groupby(['city', 'state'])['victimID'].count().to_frame()
male_by_city.head()
```

```python
print(male_by_city.columns)
male_by_city.columns = ['males']
```

```python
female_by_city = female.groupby(['city', 'state'])['victimID'].count().to_frame()
female_by_city.head()
```

```python
female_by_city.columns = ['females']
```

```python
freq_by_city = male_by_city.merge(female_by_city, how='outer', left_on=['city', 'state'], right_index=True).reset_index()
freq_by_city.isna().sum()
```

```python
freq_by_city.fillna(0, inplace=True)
freq_by_city = freq_by_city.astype({'males': 'int16', 'females': 'int16'})
freq_by_city.head()
```

```python
state_lat_lng = df_dropped[['city', 'state', 'lat', 'lng']]
state_lat_lng = state_lat_lng.drop_duplicates(subset=['city', 'state'], keep='first')
freq_by_city = freq_by_city.merge(state_lat_lng, how='inner', on=['city', 'state'])
freq_by_city.head()
```
